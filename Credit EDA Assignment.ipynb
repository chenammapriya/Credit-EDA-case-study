# Imputing the missing values 

#### Imputing the NAME_TYPE_SUITE column

```python
app_data['NAME_TYPE_SUITE'].value_counts()
```

```python
app_data['NAME_TYPE_SUITE'].mode()
```

```python
app_data['NAME_TYPE_SUITE'].fillna(value='Unaccompanied', inplace=True)
```

```python
app_data['NAME_TYPE_SUITE'].value_counts()
```

#### Imputing the AMT_REQ_CREDIT_BUREAU_ columns

```python
app_data['AMT_REQ_CREDIT_BUREAU_HOUR'].mean()
```

```python
# Filling the below columns with 0 as most of the column has 0 values in it
# AMT_REQ_CREDIT_BUREAU_HOUR
# AMT_REQ_CREDIT_BUREAU_DAY
# AMT_REQ_CREDIT_BUREAU_WEEK
# AMT_REQ_CREDIT_BUREAU_MON
# AMT_REQ_CREDIT_BUREAU_QRT
# AMT_REQ_CREDIT_BUREAU_YEAR

app_data.loc[:,'AMT_REQ_CREDIT_BUREAU_HOUR':'AMT_REQ_CREDIT_BUREAU_YEAR'] = app_data.loc[:,'AMT_REQ_CREDIT_BUREAU_HOUR':'AMT_REQ_CREDIT_BUREAU_YEAR'].fillna(0)
```

#### Imputing the AMT_GOODS_PRICE column

```python
app_data['AMT_GOODS_PRICE'].mean()
```

```python
app_data['AMT_GOODS_PRICE'].fillna(app_data['AMT_GOODS_PRICE'].mean(), inplace=True)
```

#### Imputing the AMT_ANNUITY column

```python
app_data['AMT_ANNUITY'].median()
```

```python
app_data['AMT_ANNUITY'].fillna(app_data['AMT_ANNUITY'].median(), inplace=True)
```

```python
round((app_data.isnull().sum()/len(app_data)*100),2)
```

# Creating Bins

```python
# Creating bins for income column - AMT_INCOME_TOTAL

bins_income = [0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,300000,
              325000,350000,375000,400000,425000,450000,475000,500000]
slot_income = ['0-25000','25000-50000','50000-75000','75000-100000','100000-125000','125000-150000','150000-175000',
              '175000-200000','200000-225000','225000-250000','250000-275000','275000-300000','300000-325000',
              '325000-350000','350000-375000','375000-400000','400000-425000','425000-450000','450000-500000','500000 and above']

app_data['AMT_INCOME_TOTAL_RANGE'] = pd.cut(app_data['AMT_INCOME_TOTAL'], bins_income, labels=slot_income)
```

```python
# Creating bins for income column - AMT_CREDIT

bins_credit = [0,150000,200000,250000,300000,350000,400000,450000,500000,550000,
               600000,650000,700000,750000,800000,850000,900000,950000,1000000]
slot_credit = ['0-150000','150000-200000','200000-250000','250000-300000','300000-350000','350000-400000','400000-450000',
              '450000-500000','500000-550000','550000-600000','600000-650000','650000-700000','700000-750000',
              '750000-800000','800000-850000','850000-900000','900000-950000','950000 and above']

app_data['AMT_CREDIT_RANGE'] = pd.cut(app_data['AMT_CREDIT'], bins_credit, labels=slot_credit)
```
